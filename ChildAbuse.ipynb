```python
import numpy as np 
import pandas as pd 
import tensorflow as tf 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('CSA-Data.csv')
data
```

```python
data.rename(columns={'"Children are safe among family members such as grandparents, uncles, aunts, cousins"' : 'A', 
                     '"Children are mainly abused by strangers in our society"' : 'B', 
                       'Male children dont need sexual abuse prevention knowledge' : 'C',
                       '"Teaching sexual abuse prevention in school is not necessary. It will make children curious about sex"' : 'D', 
                       'Do you know what child grooming is?' : 'E', 
                       'Do you know what signs to look for to identify if your child has been abused?' : 'F', 
                       'Do you think children need post abuse counseling for recovering?' : 'G', 
                       'Do you think you should take legal action against the abuser of your child?' : 'H', 
                       'Knowledge Level' : 'I'}, inplace = True)
```

```python
data
```

```python
lab_encode = LabelEncoder()
data_encoded = data.apply(lab_encode.fit_transform)
```

```python
data_encoded
```

```python
X = data_encoded.drop('I', axis = 1)
Y = data_encoded['I']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state = 69420)
```

```python
from tensorflow.keras.layers import Dropout

model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs = 500, batch_size = 32, validation_split = .2)
```
